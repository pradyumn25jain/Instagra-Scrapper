```python
import urllib.request
import os
import shutil
from tqdm.notebook import tqdm
from utils.scraper import Scraper
sc = Scraper()
```

```python
results = sc.get_images_from_hashtag_page("toonapp",10)
```

```python
os.mkdir("./data/AllData"))
for (date,link,name) in tqdm(results):
    try:
        path = os.path.join(f"./data/{date}")
        if not os.path.exists(path):
            os.mkdir(path)
        urllib.request.urlretrieve(
            link, f"{path}/{name}.png")
        shutil.copyfile(f"{path}/{name}.png", f"./data/AllData/{name}.png")
    except:
        print((date,link,name))
    
```
